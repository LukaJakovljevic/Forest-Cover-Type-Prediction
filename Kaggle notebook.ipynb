```python
import pandas as pd
import numpy as np
#read the training set
df = pd.read_csv('train-set.csv')
```

```python
df.head(5)
```

```python
df.info()
```

```python
#training data with Random Forest Classifier
train_data = df.values
from sklearn.ensemble import ExtraTreesClassifier
model = ExtraTreesClassifier(n_estimators=200, verbose=1, n_jobs = 4)
#The column Cover_type will be our second input
#and the set of other features as the first
model = model.fit(train_data[0:,1:55], train_data[0:,55])
```

```python
#read the test
df_test = pd.read_csv('test-set.csv')
df_test.head(5)
```

```python
test_data = df_test.values
#We now apply the trained model to the test data, omitting the column Id
output = model.predict(test_data[:,1:])
```

```python
result = np.c_[test_data[:,0].astype(int), output.astype(int)]
df_result = pd.DataFrame(result[:,0:2], columns=['Id', 'Cover_Type'])
df_result.head(5)
```

```python
df_result.to_csv('sub2.csv', index=False)
```
